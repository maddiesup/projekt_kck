```python
# wyszukujemy indeksy w których pojawia się 'dół', a wcześniejszy indeks jest 'większy'. 
mrugniecia = []
for i in range(len(df)):
    if df['signal'][i]<-40000 and df['signal'][i-1]>-40000:
        mrugniecia.append(df['time'][i])
```

```python
len(mrugniecia)
```

```python
mrugniecia_rounded=[]
for i in range(len(mrugniecia)):
    rounded = mrugniecia[i]//1
    mrugniecia_rounded.append(rounded)
mrugniecia_rounded
```

```python
litery_czas = 'litery_czas.txt'
litery = pd.read_csv(litery_czas, names = ['letter','time'])
```

```python
litery
```

```python
litery_rounded=[]
for i in range(len(litery)):
    rounded = litery['time'][i]//1
    litery_rounded.append(rounded)
litery_rounded
```

```python
wymrugane = []
for i in range(len(mrugniecia)):
    for j in range(len(litery)):
        if mrugniecia_rounded[i] == litery_rounded[j]:
            wymrugane.append(litery['letter'][j])
```
